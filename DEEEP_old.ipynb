```python
df=pd.read_csv('./pdb_features.csv')
# apply: I will apply function!
# lambda: the function name is lambda x:
# x.split('_mute')[0] if '_mute' in x: when we split word with '_mute', the first element
# else x.split('.pdb')[0] if no '_mute', split based on .pdb and export first element
df['complex_id'] = df['filename'].apply(lambda x: x.split('_mute')[0] if '_mute' in x else x.split('.pdb')[0])
df['label'] = df['filename'].apply(lambda x: 0 if '_mute' in x else 1)
gkf = GroupKFold(n_splits=5)
splits = list(gkf.split(df, groups=df['complex_id']))

# The accuracy_metric function calculates the accuracy of model predictions by comparing the predicted class
# (determined by the index of the maximum value along the class dimension) 
# with the true labels, converting the comparison to a float tensor, and averaging the 
# result to yield the proportion of correct predictions.
def accuracy_metric(inp, targ):
    return (inp.argmax(dim=1) == targ).float().mean()

all_train_losses, all_valid_losses, all_accuracies = [], [], []
all_precisions, all_recalls, all_f1s = [], [], []

for fold, (train_idx, valid_idx) in enumerate(splits):
    # \n = line change
    print(f"\nFold {fold+1}")
    # show the message immediately
    sys.stdout.flush()


    # iloc: select raw, []:select column
    print("Train groups:", df.iloc[train_idx]['complex_id'].unique())
    print("Validation groups:", df.iloc[valid_idx]['complex_id'].unique())
    sys.stdout.flush()

    # Fill missing values
    # Fille the NaN with mean of each columns
    df[cont_names] = df[cont_names].fillna(df[cont_names].mean())

    # Normalize continuous variables
    df[cont_names] = (df[cont_names] - df[cont_names].mean()) / df[cont_names].std()

    # Create the data loaders without the FillMissing and Normalize procs
    procs = [Categorify]


    dls = TabularDataLoaders.from_df(
        df, y_names='label', procs=procs,
        cont_names=cont_names, bs=16, splits=(train_idx, valid_idx)
    )

    
    # Create a tabular learner
    learn = tabular_learner(dls, layers=[200, 100], model_dir='Ha_score_model', metrics=accuracy_metric)

    # Find the optimal learning rate
    learn.lr_find()
    learn.recorder.plot_lr_find()

    # Based on the plot, set max_lr to a suitable value
    max_lr = 1e-2

    # Train the model with more epochs
    learn.fit_one_cycle(20, max_lr)

    # Plot the learning rate schedule
    learn.recorder.plot_sched()

    # Plot the loss
    learn.recorder.plot_loss()

    # Store metrics for this fold
    train_loss, valid_loss = learn.recorder.values[-1][:2]
    accuracy = learn.recorder.values[-1][2]
    
    all_train_losses.append(train_loss)
    all_valid_losses.append(valid_loss)
    all_accuracies.append(accuracy)

    # Get predictions for the validation set
    val_preds, val_labels = learn.get_preds(ds_idx=1)
    val_preds = val_preds.argmax(dim=1)

    # Calculate precision, recall, and F1-score
    precision = precision_score(val_labels, val_preds, zero_division=1)
    recall = recall_score(val_labels, val_preds, zero_division=1)
    f1 = f1_score(val_labels, val_preds, zero_division=1)

    all_precisions.append(precision)
    all_recalls.append(recall)
    all_f1s.append(f1)

    print(f"Fold {fold+1} - Precision: {precision}, Recall: {recall}, F1-score: {f1}")
    sys.stdout.flush()

# Print average results across all folds
print(f"\nAverage Training Loss: {sum(all_train_losses)/len(all_train_losses)}")
print(f"Average Validation Loss: {sum(all_valid_losses)/len(all_valid_losses)}")
print(f"Average Accuracy: {sum(all_accuracies)/len(all_accuracies)}")
print(f"Average Precision: {sum(all_precisions)/len(all_precisions)}")
print(f"Average Recall: {sum(all_recalls)/len(all_recalls)}")
print(f"Average F1-score: {sum(all_f1s)/len(all_f1s)}")
sys.stdout.flush()
```
